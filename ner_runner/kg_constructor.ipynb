```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
search_txt = """ the mount galen volcanics consists of basalt, andesite, dacite, and rhyolite lavas and dacite and rhyolite tuff and tuff-breccia """
doc = nlp(search_txt)
```

```python
for token in doc:
    print(token.i, token.text, token.pos_, token.dep_, token.head.i)
```

```python
displacy.serve(doc, style="dep")
```
